**Jawaban**
1. Matplotlib secara default menampilkan gambar dalam format RGB (Red, Green, Blue). Sementara itu, OpenCV membaca gambar dalam format BGR (Blue, Green, Red). Oleh karena itu, jika gambar yang dibaca dengan OpenCV langsung ditampilkan menggunakan Matplotlib tanpa konversi, warna akan terlihat tidak sesuai. Untuk menyesuaikan format warna, digunakan cv.cvtColor(img, cv.COLOR_BGR2RGB).

2. Perbedaan antara int16 dan int32 terletak pada kapasitas memori dan rentang nilai. Tipe data int16 lebih hemat memori dan cukup untuk menyimpan nilai piksel standar (0–255) pada citra RGB. Sedangkan int32 memiliki rentang nilai lebih besar dan cocok digunakan jika diperlukan penyimpanan nilai piksel yang lebih tinggi untuk keperluan perhitungan matematis atau pemrosesan lanjutan.

3. Fungsi cv2_imshow() adalah pengganti cv.imshow() yang dirancang khusus untuk digunakan di Google Colab. Karena Colab tidak mendukung tampilan GUI seperti di desktop, cv2_imshow() memungkinkan gambar OpenCV ditampilkan langsung di dalam notebook. Tanpa fungsi ini, hasil dari cv.imread() atau operasi seperti cv.rectangle() tidak akan terlihat.

4. skimage.io.imread() digunakan untuk membaca gambar dari URL atau file lokal. Berbeda dengan cv.imread() yang menghasilkan gambar dalam format BGR, fungsi ini langsung menghasilkan gambar dalam format RGB, sehingga lebih cocok digunakan bersama Matplotlib tanpa perlu konversi warna tambahan.
