```python
import cv2 as cv
from google. colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/PCVK/Images/kitten01.jpg')

# Warna gambar menjadi biru karena matplotlib mempunyai default format gambar BGR dan bukan RGB
# Sehingga dibutuhkan cvtColor(imgName, cv.COLOR_BGR2RGB) untuk mengubahnya
plt.imshow(img) #perhatikan hasilnya adalah citra dgn channel warna BGR
plt.show() #plt.show() akan menampilkan gambar yang sudah disiapkan (imshow())

print()

img_copy = img.copy()
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img3) #perhatikan hasilnya adalah citra dgn channel warna RGB
plt.show() #plt.show() akan menampilkan gambar yang sudah disiapkan (imshow())
```

```python
# Membaca gambar sebagai grayscale (1 channel)
img_gray = cv.imread('/content/drive/MyDrive/PCVK/Images/kitten01.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.show()

# Matplotlib tahu bahwa ini adalah citra intensitas, dan menampilkan dengan skema warna hitam-putih yang benar
plt.imshow(img_gray, cmap='gray')
plt.show()

# menampilkan citra grayscale dengan colormap artistik (gradasi ungu-kuning) untuk visualisasi yang lebih menarik
plt.imshow(img_gray, cmap='magma')
plt.show()
```

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

```python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300], [200,400]], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) # nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili R, G, dan B
pts
```

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```
